```python
import pandas as pd

Titanic=pd.read_csv("Titanic-Dataset.csv")
```

```python
Titanic
```

```python
Titanic.fillna(0)
```

```python
Titanic=Titanic.fillna(0)
```

```python
Titanic
```

```python
#converting the Sex column (categorical) column to Numerical column
#.map is one of the functions used to convert a categorical column to numerical column
#pass in a dictionary into a map function

Titanic["Sex"].map({"male":0,"female":1})

Titanic["Sex"]=Titanic["Sex"].map({"male":0,"female":1})
```

```python
#extract columns needed for prediction
#x is independent , Y is dependent

x=Titanic[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]]
y=Titanic["Survived"]
```

```python
#split the x and y

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
# fit the model(train)

from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
a=model.predict(x_train)
```

```python
# comparing x_train and prediction on x_train

xtrain_table= pd.DataFrame(x_train)
```

```python
xtrain_table["Prediction"]= ""
```

```python
xtrain_table
```

```python
xtrain_table["Prediction"]=a
```

```python
xtrain_table
```

# Good Accuracy Score

```python
#Evaluate result in logistic regression

from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_train, a)
```
